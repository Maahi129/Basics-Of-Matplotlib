```python
# Markers
import numpy as np
import matplotlib.pyplot as plt

x = np.array([10,20,30,40])
y = np.array([15,30,40,60])
plt.plot(x, y, '+')              # in place of 'o' you can use '*', 'X', 'x', 'o', 'H', '+', 'P', 'S', 'D', 'd', 'p', '>', '1', '2', '3'
plt.show()
```

### Markers Size,colors,LineStyle Description

**Color Reference**
Red          -         'r'
Green        -         'g'
Blue         -         'b'
Cyan         -         'c'
White        -         'w'
Magneta      -         'm'
Yellow       -         'y'
Black        -         'k'

MarkerEdgeColor will be represented as **mec**, 
MarkerFaceColor Will be represented as **mfc**
(You can use hex color in place of colors like 'r')

**Marker Size**
It is represented as **ms**

**Line Styles**
Dotted       -         ':'
Solid        -         '-'
Dashed       -         '--'

This are represented with **linestyle**

```python
# Example
x1 = [1,2,3]
y1 = [5,6,7]
plt.plot(x1,y1, marker='o', ms=20, mec='r', mfc='c', linestyle='--')
```

```python
# Lables , Titles , location and Fontdict

font = {'family':'serif','color':'blue','size':20}      # Font Styles

plt.title("Laptops", fontdict = font)                   # Applying Through fontdict
plt.xlabel("Our budget", loc = 'left')                  # We can specify location of labels with loc
plt.ylabel("Lap Cost", loc = 'bottom')
plt.plot(x1)
plt.plot(y1)
plt.legend(["Budget","Cost"], loc = 'lower right', bbox_to_anchor = (0.87, 0))  # with bbox we can control loc of legend 
plt.show()
```

```python
# Grids 
import numpy as np
y = np.array([1,2,3])
plt.plot(y)   # It takes x values as default and plot the graph
plt.grid( axis = 'x', color = "green", linestyle = '--', linewidth = 2) # seperately we can specify axis and styling to that grid
plt.show()
```

```python
# Ticks For x and y axis
x = [1,3,6,8,10,35,60,70]
y = [2,4,5,7,6,8,10,14]
plt.plot(x, y,'r')
x = plt.xticks(np.arange(1,100,20))
y = plt.yticks(np.arange(1,20,4))
plt.show()
```

```python
# SubPlot
# You can Draw Multiple Plots  in one figure
# It takes three arguments , first one is how much rows you want and second about columns and third is index of current plot

#Plot 1
x = np.array([0, 1, 5, 10])
y = np.array([6, 8, 9, 15])
plt.subplot(2,2,1)    # 2 row and 2 columns and it is the first plot
plt.plot(x,y)

#plot2
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.show()

#plot3
x = np.array([1, 5, 8, 12])
y = np.array([5, 8, 1, 10])
plt.subplot(2,2,3)     # it goes to 2 row for 3 plot After 2 columns are filled in 1st row
plt.plot(x,y)
plt.show()
```

```python
# Scatter Plot

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)
plt.title("Cars",color = "m")

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, c = "red")  # we can specify colors you want

plt.show()
```

```python
# Bar Plot

x = np.array(["Mahi", "Bhanu", "Chetan", "navya"])
y = np.array([14, 15, 18, 20])

plt.barh(x,y, height = 0.4, color = "red")   # we have to specify height for horizantal bars  
# You can specify width of bar
#  plt.bar(x,y, weight = 0.4, color = "red")
```

```python
# Histrogram

# It is used in frequency distributions
# For examplpe If you want to know 150 people heights , we can use this.

x = np.random.normal(150,10,200)           # (mean, standard Deviation, size of array)
plt.hist(x)
```

```python
# Pie Chart

num = [25,30,25,20]
x = ['cars', 'Bikes', 'Areoplanes','Helicopter']
my_explode = [0.25,0,0,0]
my_colors = ["red", "black","yellow","hotpink"]
plt.pie(num, labels = x, startangle = 90, explode = my_explode, shadow = True, colors = my_colors , wedgeprops={
    'edgecolor' : 'black', 'linewidth' : 1
},)
# we can give our own colors , explode will remove the part from pie which you wanna explode slightly 
plt.legend(title = "Automobile",loc = 'lower right')
plt.show()
```

```python
# Plotting two bars

plt.bar([0.25,1],[0.75,1], label = "num", width = 0.3, color = 'red')
plt.bar([0.50,1.5],[1,1.25], label = "num1", width = 0.3)
plt.legend(title = "Numbers")
plt.xlabel('num')
plt.ylabel('count')
plt.title('Counting Numbers')
plt.show()

```

```python
a = np.random.normal(80,3,180)
b = np.random.normal(100,2,200)

data = [a, b]

fig = plt.figure()

c = fig.add_axes([0,0,1,1])   # adding axes for size of the figure

bp = c.violinplot(data)
```

```python
# Box Plot

a = np.random.normal(80,3,180)
b = np.random.normal(100,2,200)

data = [a, b]

fig = plt.figure()

c = fig.add_axes([0,0,1,1])  

dp = c.boxplot(data, patch_artist = True, widths = 0.5)
```

```python
# Fill Between Graph

a = np.linspace(0, 8, 16)
b =  3 + 4*a/8 + np.random.uniform(0.0, 0.5, len(a))
e =  1 + 2*a/8 + np.random.uniform(0.0, 0.5, len(a))
fig = plt.figure()

c = fig.add_axes([0,0,1,1])  
c.fill_between(a,b,e)
```

```python
# Step Graph

x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

plt.step(x,y)
plt.show()
```

```python
# Stem graphs

x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

plt.stem(x,y)
plt.show()
```

```python
#Sine wave And Cosine wave

x = np.arange(0, 4*np.pi,0.1)
y = (np.sin(x))
z = (np.cos(x))

plt.plot(x,y,x,z)
plt.show()
```

```python
# Random Normal Plot

x = np.random.normal(1,101,100)
y = np.random.normal(1,101,100)

plt.plot(x,y)
plt.show()
```
